```python
import os
import time
import numpy as np
import csv
CSV_FILENAME = 'greeted_people.csv'
```

```python
def load_greeted_people():
    """Load the greeted people from the CSV file."""
    greeted_people = []
    if os.path.exists(CSV_FILENAME):
        with open(CSV_FILENAME, 'r') as file:
            reader = csv.reader(file)
            for row in reader:
                name = row[0]
                encoding = np.array([float(x) for x in row[1:-1]])
                last_greet_time = float(row[-1])
                greeted_people.append((name, encoding, last_greet_time))
    return greeted_people
load_greeted_people()
```

```python

def save_greeted_person(name, encoding, greet_time):
    """Save or update the greeted person's info in the CSV file."""
    greeted_people = load_greeted_people()

    # Check if person exists in CSV, update if found, append if not
    updated = False
    for i, (existing_name, existing_encoding, _) in enumerate(greeted_people):
        if np.allclose(existing_encoding, encoding, atol=0.6):
            greeted_people[i] = (name, encoding, greet_time)
            updated = True
            break

    if not updated:
        greeted_people.append((name, encoding, greet_time))

    # Write back to CSV
    with open(CSV_FILENAME, 'w', newline='') as file:
        writer = csv.writer(file)
        for name, encoding, greet_time in greeted_people:
            writer.writerow([name] + list(encoding) + [greet_time])

```
